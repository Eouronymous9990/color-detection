```python
import cv2
import pandas as pd

img_path = r"C:\Users\zbook 17 g3\Desktop\New folder\colorful-umbrellas-hanging-over-old-town-street-in-avignon-france-FOLF04341.jpg"
img = cv2.imread(img_path)

# Preparing csv data for the algorithm
columns_names = ["color", "color_name", "hex", "R", "G", "B"]
colors = pd.read_csv(r"C:\Users\zbook 17 g3\Downloads\colors(1).csv", names=columns_names, header=None)

# Calculate the coordinates by mouse click and identify b,g,r values
click = False
def distance_color(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b, g, r, x_pos, y_pos, click
        x_pos = x
        y_pos = y
        b, g, r = img[y, x]
        b = int(b)
        g = int(g)
        r = int(r)
        click = True

# Calculate which closest color to the unknown color
def calculate_coordinates(r_, g_, b_):
    minimum_distance = 10000
    colsname = "Unknown"  # Default value for when no match is found
    for i in range(len(colors)):
        rdiff = abs(r_ - int(colors.loc[i, "R"]))
        gdiff = abs(g_ - int(colors.loc[i, "G"]))
        bdiff = abs(b_ - int(colors.loc[i, "B"]))
        distance = rdiff + gdiff + bdiff
        if distance < minimum_distance:
            minimum_distance = distance
            colsname = colors.loc[i, "color_name"]  # Get the color name with the minimum distance
    return colsname

cv2.namedWindow('image')
cv2.setMouseCallback('image', distance_color)

while 1:
    cv2.imshow("image", img)
    if click:
        # Rectangle to display the text
        cv2.rectangle(img, (40, 20), (800, 80), (b, g, r), -1)

        text = calculate_coordinates(r, g, b)  # Color name
        
        if b+g+r>600:
            cv2.putText(img, text, (50, 50), 2, 1, (0, 0, 0), 2, cv2.LINE_AA)# to handle bright colors
        else:
            cv2.putText(img, text, (50, 50), 2, 1, (255, 255, 255), 2, cv2.LINE_AA)    
            
            

        click = False

    # Break the loop
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
